
# Ollama Model Testing

Test local Ollama models.



```python

import requests

models = [
    "llama3:latest",
    "mistral:latest",
    "phi3:latest",
    "deepseek-coder:latest"
]

for model in models:
    payload = {
        "model": model,
        "prompt": "Explain Python in one sentence",
        "stream": False
    }

    response = requests.post(
        "http://localhost:11434/api/generate",
        json=payload,
        timeout=120
    )

    print(model)
    print(response.json().get("response"))
    print("-" * 50)

```


# Career AI Chat Demo



```python

import requests

payload = {
    "resume": "Python backend engineer",
    "jobDescription": "Backend AI engineer",
    "modelName": "llama3:latest",
    "userMessage": "How can I improve my resume?"
}

response = requests.post(
    "http://localhost:8000/chat-assistant",
    json=payload
)

print(response.json())

```
